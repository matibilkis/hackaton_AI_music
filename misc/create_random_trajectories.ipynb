```python
import numpy as np
import matplotlib.pyplot as plt 
from IPython.display import display, HTML

class Record:
    def __init__(self, T_0):
        self.transitions = [T_0]
        self.trajectories = []
        
    def give_transition_matrix(self,melody):
        
        lookup = np.empty((5))
        lookup[[0,1,2,3,4]] = np.arange(5)
        counts = np.zeros((5,5))
        np.add.at(counts, (melody[:-1], melody[1:]),1 )
        probs = counts/counts.sum(axis=0, keepdims=True)
        self.transitions.append(probs)
        return probs
    

notes_to_choose = 5
melody_duration = 1000
number_of_melodies = 15


T_matrix = np.ones((notes_to_choose,notes_to_choose))/notes_to_choose
```

```python
first_note_ever = 0
melodies = []

record_obj = Record(T_matrix)


state = first_note_ever
for it in range(number_of_melodies):
    melody=[]
    for k in range(melody_duration):
        melody.append(state)
        state = np.random.choice(range(notes_to_choose),1, p=T_matrix[:,state])[0]
    T_matrix = record_obj.give_transition_matrix(melody)
    record_obj.trajectories.append(melody)
```

```python
for k in record_obj.transitions:
    plt.imshow(k)
    plt.show()
```

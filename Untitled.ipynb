```python
def z(x,y, p, ini=2):
    image_array = 0
    for indi,k in zip(range(ini,ini+ len(p)+1), p ):
        image_array+= k*np.sin(indi*x*np.pi/2)*np.sin(indi*y*np.pi/2)
    return image_array
```
